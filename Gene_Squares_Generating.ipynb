```python
import pandas as pd
import numpy as np
import itertools
from sklearn import preprocessing
```

```python
%%time
tcga_gene_map=pd.read_hdf("F:/MyData/Pathway_Attention/data/hugo_ensembl_geneExpre_map.h5",key='tcga_gene_map')
geo_gene_map=pd.read_hdf("F:/MyData/Pathway_Attention/data/hugo_ensembl_geneExpre_map.h5",key='geo_gene_map')
```

```python
pathways=pd.read_csv('F:/MyData/Pathway_Attention/data/All_Sub-pathways/allPathways.csv',sep='\,',engine='python',header=None,index_col=0)
pathways
```

```python
pathways=pathways.fillna(0)
pathways
```

```python
pathways=pathways.drop(pathways.index[40])
```

```python
pathways
```

```python
tcga_col=tcga_gene_map.columns.tolist()
```

```python
tcga_col.pop(0)
```

```python
tcga_col.pop(0)
```

```python
tcga_col
```

```python
%%time
from tqdm import tqdm
import time
for sample in tqdm(tcga_col):
    #time.sleep(0.25)
    expre_mat=np.empty((40,46))
    for i in range(len(pathways)):
        pathway=np.array(pathways.iloc[i]).tolist()
        path_expre=tcga_gene_map[tcga_gene_map['geneName'].isin(pathway)][sample].tolist()
        while len(path_expre)!=46:
            if len(path_expre)<46:
                path_expre.append(0)
        expre_mat[i]=path_expre
    list49=[0 for m in range(49)]
    list46=[0 for n in range(46)]
    for p in range(9):
        expre_mat=np.row_stack((expre_mat,list46))
    for q in range(3):
        expre_mat=np.column_stack((expre_mat,list49))
    expre_df=pd.DataFrame(expre_mat).T
    chunk1=expre_df.iloc[:,0].values.reshape(7,7)
    chunk2=expre_df.iloc[:,7].values.reshape(7,7)
    chunk3=expre_df.iloc[:,14].values.reshape(7,7)
    chunk4=expre_df.iloc[:,21].values.reshape(7,7)
    chunk5=expre_df.iloc[:,28].values.reshape(7,7)
    chunk6=expre_df.iloc[:,35].values.reshape(7,7)
    chunk7=expre_df.iloc[:,42].values.reshape(7,7)
    for c in range(1,7):
        block=expre_df.iloc[:,c].values.reshape(7,7)
        chunk1=np.concatenate([chunk1,block],axis=1)
    for c in range(8,14):
        block=expre_df.iloc[:,c].values.reshape(7,7)
        chunk2=np.concatenate([chunk2,block],axis=1)
    for c in range(15,21):
        block=expre_df.iloc[:,c].values.reshape(7,7)
        chunk3=np.concatenate([chunk3,block],axis=1)
    for c in range(22,28):
        block=expre_df.iloc[:,c].values.reshape(7,7)
        chunk4=np.concatenate([chunk4,block],axis=1)
    for c in range(29,35):
        block=expre_df.iloc[:,c].values.reshape(7,7)
        chunk5=np.concatenate([chunk5,block],axis=1)
    for c in range(36,42):
        block=expre_df.iloc[:,c].values.reshape(7,7)
        chunk6=np.concatenate([chunk6,block],axis=1)
    for c in range(43,49):
        block=expre_df.iloc[:,c].values.reshape(7,7)
        chunk7=np.concatenate([chunk7,block],axis=1)
    pathway_blocks=np.concatenate([chunk1,chunk2,chunk3,chunk4,chunk5,chunk6,chunk7],axis=0)
    np.save('data/pathway_block_images/tcga_all/{}.npy'.format(sample),pathway_blocks)
```
